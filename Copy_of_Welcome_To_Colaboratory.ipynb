```
import numpy as np
import matplotlib.pyplot as plt

# Define a time period (e.g., years)
years = np.arange(2000, 2021)  # Replace with your desired time range

# Replace the following synthetic data with actual demographic data for Laayoune
population = np.random.randint(10000, 50000, len(years))
birth_rate = np.random.uniform(10, 20, len(years))
death_rate = np.random.uniform(5, 15, len(years))

# Create subplots
fig, ax1 = plt.subplots()

# Plot demographic challenges
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color='tab:blue')
ax1.plot(years, population, color='tab:blue', label='Population')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_ylim(0, max(population) + 5000)  # Adjust the y-axis limits

# Create a second y-axis to plot other demographic challenges
ax2 = ax1.twinx()
ax2.set_ylabel('Rates', color='tab:red')
ax2.plot(years, birth_rate, color='tab:red', label='Birth Rate', linestyle='--')
ax2.plot(years, death_rate, color='tab:orange', label='Death Rate', linestyle='--')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.set_ylim(0, max(birth_rate + death_rate) + 5)  # Adjust the y-axis limits

# Add a legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Set the title and labels
plt.title('Demographic Challenges in Laayoune Over Time')
plt.xlabel('Year')

# Display the plot
plt.tight_layout()
plt.show()
```
